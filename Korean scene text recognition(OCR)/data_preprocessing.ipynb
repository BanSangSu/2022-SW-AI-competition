```python
# ./variable_korea_ocr Data import Processing
'''
    variable_korea_ocr Folder에 속한 label json File을 txt File로 변환
    letter -> json File의 "info" -> "text"
    word -> json File의 "text" -> "word" -> "value"
'''
import os
# path = r'E:\Kim\dacon_ocr_competition\data\variable_korea_ocr'
path = './data/variable_korea_ocr'
path_ = './variable_korea_ocr'
# os.chdir(path)
import json

data_names = ['letter', 'word']

list = ['Training_label', 'validation_label']

# with open(path+'\data.txt', 'w', encoding="UTF-8") as f:
with open(path+'/data.txt', 'w', encoding="UTF-8") as f:
    for folder in list:
        label_list = os.path.join(path, folder)
        for i in data_names:
            label_list_ = os.path.join(label_list, i)
            label_list_num = os.listdir(label_list_)
            if i == 'letter':
                for n in label_list_num:
                    for k in os.listdir(os.path.join(label_list_, n)):
                        with open(os.path.join(label_list_, n, k), 'r', encoding="UTF-8") as j:
                            json_data = json.load(j)
                            v0 = json_data['info']['text']
                            v1 = json_data['image']['file_name']
                            if folder == 'validation_label':
                                f.write(path_+'/validation/'+i+'/'+n+'/'+v1+'\t'+v0+'\n')
                            elif folder == 'Training_label':
                                f.write(path_+'/Training/'+i+'/'+n+'/'+v1+'\t'+v0+'\n')
            elif i == 'word':
                for n in label_list_num:
                    for k in os.listdir(os.path.join(label_list_, n)):
                        with open(os.path.join(label_list_, n, k), 'r', encoding="UTF-8") as j:
                            json_data = json.load(j)
                            v0 = None
                            for j in range(len(json_data['text']['word'])):
                                if j == 0: v0 = json_data['text']['word'][j]['value']
                                else: v0 += json_data['text']['word'][j]['value']
                            v1 = json_data['image']['file_name']
                            if folder == 'validation_label':
                                f.write(path_+'/validation/'+i+'/'+n+'/'+v1+'\t'+v0+'\n')
                            elif folder == 'Training_label':
                                f.write(path_+'/Training/'+i+'/'+n+'/'+v1+'\t'+v0+'\n')
```

### Make characterset

```python
characterset = set()
with open("./valid.txt", "r", encoding="UTF-8") as f:
    for line in f:
        tab_index = line.find("\t")
        next_line_index = line.find("\n")
        label = line[(tab_index+1):next_line_index]
        for character in label:
            if character != " ":
                characterset.add(character)
        # print(label)
        
# print(" " in characterset)
characterset = list(characterset)
sorted_characterset = sorted(characterset, key=lambda characterset: characterset[0])
print(len(sorted_characterset))

model_characterset = ""
for character in sorted_characterset:
    model_characterset += character
# 이거 복사해서 쓸까용?ㅋㅋㅋㅋㅋ
print(model_characterset)
```

```python
#!-*- coding: utf-8 -*-

def nextKorLetterFrom(letter):
    lastLetterInt = 15572643

    if not letter:
        return '가'

    a = letter
    b = a.encode('utf8')
    c = int(b.hex(), 16)

    if c == lastLetterInt:
        return False

    d = hex(c + 1)
    e = bytearray.fromhex(d[2:])

    flag = True
    while flag:
        try:
            r = e.decode('utf-8')
            flag = False
        except UnicodeDecodeError:
            c = c + 1
            d = hex(c)
            e = bytearray.fromhex(d[2:])

    return e.decode()

def test():

    flag = True
    k = ''

    while flag:
        k = nextKorLetterFrom(k)
        with open('output.txt', 'a') as out:
            if k is not False:
                out.write(k)
                # print(k + ',')
        if k is False:
            flag = False
test()
```
